# Aprendizado não Supervisionado

O **aprendizado não supervisionado** é um tipo de técnica de **aprendizado de máquina** onde um modelo aprende a identificar padrões e estruturas em dados que não possuem rótulos ou respostas previamente definidas. Ao contrário do aprendizado supervisionado, que trabalha com pares de entrada e saída, o aprendizado não supervisionado lida apenas com dados de entrada, buscando agrupar ou organizar as informações de maneira significativa.

### Como funciona?

No aprendizado não supervisionado, o algoritmo recebe um grande volume de dados brutos e é encarregado de extrair informações úteis sem qualquer orientação explícita. A ausência de rótulos nos dados significa que o modelo deve encontrar padrões ocultos por conta própria. O objetivo é explorar a estrutura subjacente dos dados, suas distribuições ou relacionamentos.

#### Principais Técnicas de Aprendizado Não Supervisionado:

1. **Agrupamento (Clustering)**:
   A técnica mais comum, onde o objetivo é dividir os dados em grupos (ou clusters) baseados em semelhanças.
   - **K-Means**: Um dos métodos mais populares de agrupamento, onde os dados são particionados em \( k \) clusters baseados na proximidade dos pontos em relação aos centróides.
   - **DBSCAN**: Agrupa pontos densamente conectados e ignora ruído ou outliers, sendo útil em conjuntos de dados com formas complexas.

2. **Redução de Dimensionalidade**:
   Usada para reduzir o número de variáveis (ou dimensões) de um conjunto de dados, mantendo suas características mais importantes.
   - **PCA (Análise de Componentes Principais)**: Um método que projeta dados de alta dimensionalidade em um espaço de menor dimensão, maximizando a variação entre os dados.
   - **t-SNE**: Técnica de visualização usada para reduzir dados multidimensionais para duas ou três dimensões, preservando a estrutura local dos dados.

3. **Modelos de Mistura (Mixture Models)**:
   O aprendizado não supervisionado pode ser usado para estimar distribuições de probabilidade em dados.
   - **GMM (Modelos de Mistura Gaussiana)**: Assume que os dados são gerados por uma combinação de distribuições gaussianas, sendo útil para capturar variações de padrões mais complexos.

4. **Redes Neurais Autoencoders**:
   Um tipo de rede neural usada para aprender representações compactas dos dados. O objetivo é comprimir os dados de entrada em uma codificação de menor dimensão e, em seguida, reconstruir os dados originais a partir dessa codificação.
   - **Autoencoders Variacionais (VAE)**: Uma variação que gera novos dados a partir das codificações latentes aprendidas.

### Aplicações do Aprendizado Não Supervisionado:

- **Agrupamento de clientes**: Empresas podem agrupar clientes com base em seus comportamentos de compra, permitindo campanhas de marketing mais direcionadas.
- **Detecção de anomalias**: Identificar padrões anômalos em dados, como fraudes em transações financeiras ou falhas em sistemas de máquinas.
- **Segmentação de imagens**: Usado na análise de imagens para agrupar pixels semelhantes ou detectar padrões visuais.
- **Compressão de dados**: Redução de dimensionalidade é usada em compressão de dados, facilitando o armazenamento e processamento.

### Desafios:

Embora o aprendizado não supervisionado tenha grande potencial, também traz desafios, como:
- **Interpretação dos resultados**: Sem rótulos, pode ser difícil avaliar a qualidade dos grupos ou padrões encontrados.
- **Escolha de hiperparâmetros**: Métodos como o K-Means exigem a escolha do número de clusters, o que pode não ser trivial.
- **Complexidade dos dados**: Dados complexos ou ruidosos podem dificultar a extração de padrões claros e úteis.

### Conclusão:

O aprendizado não supervisionado é uma ferramenta poderosa para descobrir insights ocultos em dados sem a necessidade de rótulos ou supervisão. Ele é amplamente utilizado em áreas como processamento de linguagem natural, análise de imagens e segmentação de mercado. Embora ainda desafie pesquisadores e profissionais em termos de interpretação e aplicabilidade, suas técnicas continuam evoluindo, abrindo novas possibilidades de descoberta automática em grandes volumes de dados.

# O algorítmo KMeans

O **K-Means** é um algoritmo de **aprendizado não supervisionado** usado para resolver problemas de **agrupamento (clustering)**. Ele busca particionar um conjunto de dados em \(k\) grupos (clusters), onde \(k\) é um valor definido pelo usuário. O objetivo do K-Means é organizar os dados de forma que os pontos dentro de um mesmo grupo sejam mais próximos uns dos outros, e os grupos, por sua vez, sejam distintos entre si.

### Como o K-Means funciona?

O K-Means opera de forma iterativa para minimizar a variabilidade dentro dos clusters e maximizar a diferença entre os clusters. Aqui está um resumo de seu funcionamento:

1. **Escolha o número de clusters (\(k\))**: O valor de \(k\) é o número de grupos que o algoritmo formará. Este número é definido previamente pelo usuário.
   
2. **Inicialização dos centróides**: O algoritmo escolhe \(k\) pontos iniciais chamados de **centróides** (ou centros de clusters). Esses centróides podem ser selecionados aleatoriamente entre os dados ou definidos de maneira estratégica (como no algoritmo K-Means++).

3. **Atribuição dos pontos aos clusters**: Cada ponto de dados é atribuído ao cluster cujo centróide está mais próximo, geralmente com base em uma medida de distância, como a **distância Euclidiana** (a distância reta entre dois pontos).

4. **Recalcular os centróides**: Após a atribuição dos pontos aos clusters, o algoritmo recalcula os centróides de cada grupo, reposicionando-os para o centro de todos os pontos atribuídos ao cluster.

5. **Repetição do processo**: Atribuição e recalculação são repetidas até que os centróides parem de mudar ou até que o número máximo de iterações seja atingido. Esse processo de repetição continua até que os clusters sejam estáveis.

### Exemplo passo a passo:

Imagine um conjunto de dados bidimensional (com duas características) com 10 pontos dispersos:

1. Você define que deseja 3 clusters (\(k = 3\)).
2. O algoritmo escolhe 3 pontos aleatórios como centróides iniciais.
3. Cada ponto de dados é atribuído ao cluster cujo centróide está mais próximo.
4. O centróide de cada grupo é recalculado como o "ponto médio" dos dados pertencentes ao grupo.
5. O processo de atribuição e recalculação continua até que os clusters não mudem mais.

### Função Custo: Minimização da Soma dos Quadrados

O K-Means tenta minimizar a **soma dos quadrados das distâncias (SSE - Sum of Squared Errors)** de cada ponto ao centróide de seu cluster. Isso significa que o algoritmo busca minimizar a variabilidade dentro de cada cluster.

A função custo que o algoritmo tenta otimizar é a seguinte:

$$
J = \sum_{i=1}^{k} \sum_{x \in C_i} \| x - \mu_i \|^2
$$


Onde:
- \(x\) é um ponto de dado.
- \($\mu_i\$) é o centróide do \(i\)-ésimo cluster.
- \(C_i\) é o conjunto de pontos atribuídos ao cluster \(i\).
-∥x−μi ∥ representa a distância Euclidiana entre o ponto \(x\) e o centróide μi.

### Vantagens do K-Means:

- **Simplicidade**: O algoritmo é fácil de entender e implementar.
- **Escalabilidade**: Ele é eficiente para grandes conjuntos de dados.
- **Rapidez**: O K-Means converge rapidamente em muitas aplicações práticas.

### Desvantagens do K-Means:

- **Escolha de \(k\)**: O valor de \(k\) deve ser definido previamente e pode ser difícil escolher o número certo de clusters. Uma abordagem comum é usar o **método do cotovelo (Elbow Method)** para encontrar um valor adequado.
- **Sensível à inicialização**: A escolha dos centróides iniciais pode afetar o resultado final. A variação **K-Means++** ajuda a lidar com esse problema, escolhendo centróides iniciais de forma mais inteligente.
- **Forma dos clusters**: O K-Means assume que os clusters são esféricos e têm o mesmo tamanho, o que pode não ser adequado para dados mais complexos.
- **Sensível a outliers**: Pontos de dados anômalos podem afetar significativamente os clusters.

### Aplicações do K-Means:

- **Segmentação de clientes**: Agrupar clientes com base em seus comportamentos ou características.
- **Compressão de imagens**: Usado para reduzir a quantidade de cores em uma imagem, agrupando pixels semelhantes.
- **Análise de dados de sensores**: Detectar padrões em dados coletados por sensores.
- **Agrupamento de documentos**: Organizar documentos em grupos baseados em similaridades de conteúdo.

### Método do Cotovelo:

O **método do cotovelo** é uma técnica usada para determinar o número ideal de clusters (\(k\)) no K-Means. Ao plotar a soma dos quadrados das distâncias (SSE) para diferentes valores de \(k\), você observará uma diminuição acentuada inicialmente, seguida por uma estabilização. O ponto onde a curva começa a "achatar", formando um "cotovelo", é considerado o número ideal de clusters.

### Conclusão:

O K-Means é uma abordagem simples e eficaz para agrupar dados em grupos significativos. Ele é amplamente utilizado em diversas áreas e oferece bons resultados, especialmente em cenários onde os clusters possuem formas aproximadamente esféricas. Contudo, sua eficácia depende da escolha adequada de \(k\) e dos centróides iniciais, além de ser sensível a outliers e ruído nos dados.

```python
from sklearn.datasets import make_blobs
import pandas as pd
import numpy as np
```

```python
dataset, classes = make_blobs(n_samples=600, n_features=2, centers=4, cluster_std=.7, random_state=0)
# Adding column to numpy array
datasetWithClasses = np.column_stack((dataset, classes))
# pandas data frame visualization
df = pd.DataFrame(datasetWithClasses, columns=['var1', 'var2', 'class'])
#df = pd.DataFrame( dataset, columns=['var1', 'var2'])
# Take a look at first ten samples
df
```

```python
from matplotlib import pyplot as plt
df.plot(kind='scatter', x='var1', y='var2', s=10, alpha=.8)
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
# Conta a quantidade de amostra de cada classe
from collections import Counter
Counter(classes)
```

```python
import matplotlib.pyplot as plt
plt.scatter(dataset[:, 0], dataset[:, 1], s=30, alpha=.5)
plt.show()
```

```python
# Other way to plot using Seaborn, with classes discrimination
import seaborn as sns
sns.scatterplot(x=dataset[:, 0], y=dataset[:, 1], hue=classes, s=30)
plt.show()
```

```python
from sklearn.cluster import KMeans
from yellowbrick.cluster import KElbowVisualizer
model = KMeans(n_init=20)
visualizer = KElbowVisualizer(model, k=(1,12)).fit(df)
visualizer.show();
```

```python
from sklearn.cluster import KMeans
#kmeans = KMeans(n_clusters=4, init='k-means++', n_init=10, random_state=0).fit(df)
kmeans = KMeans(n_clusters=4, init='k-means++', n_init=10, random_state=0)
# fit kMeans with df pandas dataset
kmeans.fit(df)
```

```python
predict_classes = kmeans.labels_  # same as kmeans.predict(df)
predict_classes[0:100]
```

```python
kmeans.inertia_
```

```python
kmeans.n_iter_
```

```python
predict_centers = kmeans.cluster_centers_
predict_centers
```

```python
Counter(kmeans.labels_)
```

```python
import seaborn as sns
sns.scatterplot(data=df, x="var1", y="var2", hue=predict_classes)
plt.show()
```

```python
sns.scatterplot(data=df, x="var1", y="var2", hue=predict_classes)
plt.scatter(predict_centers[:,0], predict_centers[:,1], marker="X", c="r", s=150, alpha=0.7, label="centroids")
plt.legend()
plt.show()
```
