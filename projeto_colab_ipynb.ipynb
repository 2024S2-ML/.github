```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs

# Nmero de amostras e nmero de caractersticas
n_samples = 1000
n_features = 2
n_clusters = 2

# Desvios padro dos clusters
std_devs = [1, 2, 3]

# Criar uma figura para visualizao
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

for ax, std_dev in zip(axes, std_devs):
    # Gerar dados com desvio padro especificado
    X, y = make_blobs(n_samples=n_samples, centers=n_clusters, cluster_std=std_dev, n_features=n_features, random_state=42)

    # Plotar os dados
    ax.scatter(X[:, 0], X[:, 1], c=y, s=10, cmap='viridis', alpha=0.6)
    ax.set_title(f'Cluster Std Dev = {std_dev}')
    ax.set_xlabel('Feature 1')
    ax.set_ylabel('Feature 2')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.preprocessing import MinMaxScaler, normalize

# Funo para plotar dados
def plot_data(X, title, ax):
    ax.scatter(X[:, 0], X[:, 1], c='blue', s=10, alpha=0.6)
    ax.set_title(title)
    ax.set_xlabel('Feature 1')
    ax.set_ylabel('Feature 2')

# Gerar dados
n_samples = 1000
n_features = 2
n_clusters = 2
X, y = make_blobs(n_samples=n_samples, centers=n_clusters, cluster_std=1.5, n_features=n_features, random_state=42)

# Criar uma figura para visualizao
fig, axes = plt.subplots(2, 2, figsize=(14, 12))

# Plotar dados originais
plot_data(X, 'Original Data', axes[0, 0])

# Normalizao Normal (Z-Score)
X_normalized = (X - X.mean(axis=0)) / X.std(axis=0)
plot_data(X_normalized, 'Z-Score Normalized Data', axes[0, 1])

# Min-Max Scaling
scaler = MinMaxScaler()
X_minmax = scaler.fit_transform(X)
plot_data(X_minmax, 'Min-Max Scaled Data', axes[1, 0])

# L1 Normalization
X_l1 = normalize(X, norm='l1')
plot_data(X_l1, 'L1 Normalized Data', axes[1, 1])

# L2 Normalization
X_l2 = normalize(X, norm='l2')
plot_data(X_l2, 'L2 Normalized Data', axes[1, 1])  # Plot L2 normalization on the same subplot

plt.tight_layout()
plt.show()

```
